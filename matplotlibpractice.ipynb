```python
xweightlbs = np.array([2600,3300,5000,7500,2000,4000,4500,15000])
ytopspeedmph = np.array([132,161,120,110,125,178,172,95])
```

```python
plt.plot(xweightlbs, ytopspeedmph,'o')
plt.title("Car Data", loc = 'left')
plt.xlabel("Car Weight (lbs)")
plt.ylabel("Top Speed (mph)")
plt.grid()
plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3,4,5])
y = np.array([129, 149, 169, 109,119,219])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y = np.array([119, 229, 149, 179, 129, 189, 199,99])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Price Data", loc = 'right')
plt.show()
```
